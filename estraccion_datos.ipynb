```python
def crear_df_partidos(equipos,dias_fechas,info_partidos,horas):
    """
    Crea un DataFrame con partidos organizados en filas y equipos en columnas.

    :param textos: Lista de equipos.
    :return: DataFrame con partidos.
    """
    # Obtener el tamaño de la lista
    tamaño_lista = len(equipos)

    # Dividir el tamaño entre 2
    resultado = int(tamaño_lista / 2)

    # Crear una lista para almacenar los datos del DataFrame
    datos = []


    # Iterar y tomar pares de elementos
    for i in range(resultado):
        # Calcular el índice de inicio para el par actual
        start_index = i * 2
        # Tomar los dos elementos
        par = equipos[start_index:start_index + 2]

        info_partido = info_partidos[i]
        hora = horas[i]

        dia_fecha = dias_fechas[i]

        # Dividir la cadena en partes usando la coma como separador
        partes = dia_fecha.split(',')

        # Extraer el texto antes de la coma
        dia = partes[0].strip()

        fecha = partes[1].strip()

        # Crear una nueva variable en el formato numero_dia/mes/2024
        fecha_partidos = f"{fecha}/2024"
        
        # Crear un diccionario para cada partido
        partido = {
            'equipo1': par[0],
            'equipo2': par[1],
            'fecha': fecha_partidos,
            'dia_de_la_semana':dia,
            'info_partido':info_partido,
            'hora':hora
        }

        datos.append(partido)

    # Crear el DataFrame
    df = pd.DataFrame(datos)
    return df
```

```python
import re
import sqlite3
import pandas as pd

def insertar_datos_en_sqlite(db_name, table_name, dataframe):
    """
    Inserta los datos de un DataFrame en una tabla SQLite.

    :param db_name: Nombre de la base de datos SQLite.
    :param table_name: Nombre de la tabla en la base de datos.
    :param dataframe: DataFrame con los datos a insertar.
    """
    # Conectar a la base de datos SQLite
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()

    # Iniciar una transacción
    cursor.execute('BEGIN TRANSACTION')

    # Función para encerrar entre comillas los nombres de columnas que lo necesiten
    def quote_identifier(s):
        return f'"{s}"' if re.match(r'^[0-9]', s) or ' ' in s or s.upper() in ['TO', 'FROM', 'WHERE', 'GROUP', 'ORDER'] else s

    # Crear una lista de nombres de columnas con los identificadores apropiadamente citados
    columnas = [quote_identifier(col) for col in dataframe.columns]

    # Crear la parte de los valores de la consulta SQL
    valores = ', '.join(['?' for _ in dataframe.columns])

    # Crear la consulta SQL
    sql = f"INSERT INTO {table_name} ({', '.join(columnas)}) VALUES ({valores})"

    # Iterar sobre las filas del DataFrame
    for _, fila in dataframe.iterrows():
        # Ejecutar la consulta SQL con los valores de la fila
        cursor.execute(sql, tuple(fila))

    # Confirmar la transacción
    conn.commit()

    # Cerrar la conexión
    conn.close()
```
